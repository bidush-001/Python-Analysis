## Student well-being, mental health, and academic success


### Introduction
Student well-being and mental health can have a significant impact on academic success. This project analyzes a university student survey dataset on mental health and academic performance to uncover insights for improving student outcomes. The dataset contains 87 student responses and 21 features, including demographic information, CGPA (cumulative grade point average), and self-reported mental health indicators (e.g. depression, anxiety, academic pressure, financial concerns). We will perform data cleaning, exploratory analysis, clustering, and basic modeling (regression and classification) to understand the relationships between mental health factors and academic performance. The goal is to present clear, actionable findings for university stakeholders interested in student well-being and success.

### Data Preparation
First, we load the dataset and prepare it for analysis. Data preparation steps include:
#### Cleaning and Missing Values: Check for missing entries and handle them appropriately.
#### Feature Engineering: Convert CGPA ranges to numeric values and segment into performance bands (Low, Medium, High).
#### Encoding Categoricals: Convert categorical text fields (gender, major, etc.) into numeric codes for analysis.
#### Finalize Dataframe: Ensure the data is clean and ready for exploration and modeling.

Below, we execute these steps. CGPA was originally recorded in ranges (e.g. "3.5-4.0"); we map these to their midpoint (e.g. 3.75) for numerical analysis. We then create a new column cgpa_band to categorize performance as Low (CGPA < 2.5), Medium (2.5–3.5), or High (3.5–4.0). We also convert the average_sleep range into a numeric average of hours and fill any missing sleep values with a reasonable default. Finally, we encode categorical features using label encoding.

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv('MentalHealthSurvey new.csv')

# Map CGPA ranges to numeric midpoints for analysis
cgpa_mapping = {
    '0.0-0.0': 0.0,
    '1.5-2.0': 1.75,
    '2.0-2.5': 2.25,
    '2.5-3.0': 2.75,
    '3.0-3.5': 3.25,
    '3.5-4.0': 3.75
}
df['cgpa'] = df['cgpa'].map(cgpa_mapping)

# Segment CGPA into performance bands
df['cgpa_band'] = pd.cut(df['cgpa'], bins=[-np.inf, 2.5, 3.5, np.inf], labels=['Low', 'Medium', 'High'])

# Convert average_sleep ranges to numeric hours (midpoints) and handle missing values
sleep_mapping = {'2-4 hrs': 3, '4-6 hrs': 5, '7-8 hrs': 7.5}
df['average_sleep'] = df['average_sleep'].map(sleep_mapping)
df['average_sleep'] = df['average_sleep'].fillna(6)  # fill missing with ~6 hours as a reasonable default

# Encode categorical variables into numeric labels
from sklearn.preprocessing import LabelEncoder
categorical_cols = ['gender', 'university', 'degree_level', 'degree_major', 'academic_year',
                    'residential_status', 'campus_discrimination', 'sports_engagement', 'stress_relief_activities']
le = LabelEncoder()
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

# Verify the data preparation
print("Dataset shape:", df.shape)
print("CGPA band distribution:\n", df['cgpa_band'].value_counts())
print("\nSample of cleaned data:\n", df[['cgpa','cgpa_band','academic_pressure','financial_concerns','social_relationships','depression','anxiety']].head(5))
```

#### The output shows the dataset has no missing values after cleaning, and confirms the CGPA band distribution (e.g., the count of students in Low, Medium, High categories). In this sample, we observe that the data is now all numeric except the cgpa_band labels, which will be useful for analysis and visualization. Most students fall into the Medium performance band (CGPA between 2.5–3.5), followed by High and then Low. This shows a relatively balanced distribution, with a slight concentration around mid-level performers.

### Exploratory Data Analysis (EDA)
#### In this section, we explore the data to uncover patterns and relationships between academic performance (CGPA) and mental health factors. We focus on three aspects of interest for stakeholders:
#### 1: Academic Performance vs Mental Health: Do students with different performance levels report different mental health symptom levels?
#### 2: Academic Pressure Distribution: How does perceived academic pressure vary among high-performing and low-performing students?
#### 3: Correlation of Key Factors: What is the correlation between academic success, academic pressure, and financial stress?

### 1. Academic Performance and Mental Health Symptoms
#### We compare the average mental health symptom level across the CGPA performance bands. Here, we use depression level as a representative mental health symptom (rated on a scale from 1 = low to 5 = high) to see if academic performance groups differ in reported depression. We plot a bar chart of the average depression score for Low, Medium, and High CGPA groups.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Bar chart: average depression level by CGPA performance band
plt.figure(figsize=(6,4))
sns.barplot(x='cgpa_band', y='depression', data=df, estimator=np.mean, ci=None, palette='Blues')
plt.title('Average Depression Level by Performance Band')
plt.xlabel('CGPA Performance Band')
plt.ylabel('Average Depression Score (1-5)')
plt.ylim(0, 5)
plt.show()
```

#### The bar chart above shows the average depression scores for each CGPA band (Low, Medium, High). Students in the Low CGPA band report the highest levels of depression on average, whereas students in the High CGPA band have the lowest average depression scores. Medium performers fall in between. This suggests a clear pattern: academically struggling students tend to experience more severe depressive symptoms, while high-performing students report fewer depressive symptoms. For university staff, this highlights a potential link between academic difficulties and mental health, indicating that students with lower grades might need additional mental health support.

```python
# Box plot: academic pressure distribution by CGPA band
plt.figure(figsize=(6,4))
sns.boxplot(x='cgpa_band', y='academic_pressure', data=df, palette='Oranges')
plt.title('Academic Pressure by Performance Band')
plt.xlabel('CGPA Performance Band')
plt.ylabel('Academic Pressure (1-5)')
plt.show()
```

#### The box plot illustrates the spread of academic pressure scores for each performance group. We observe that academic pressure is a common experience across all performance bands. The median pressure (central line in each box) is relatively high and comparable for all groups, and the interquartile ranges overlap significantly. This means that both high-performing students and those with lower grades report feeling substantial academic pressure. In other words, academic stress affects students of all performance levels. There isn’t a drastic difference in pressure – many top students feel pressure to maintain their grades, while lower-performing students feel pressure to improve. For stakeholders, this finding emphasizes the importance of providing stress management resources to all students, not just those who are struggling academically.

### 3. Correlation Among Performance, Pressure, and Financial Stress
#### Financial stress is another key factor in student well-being. We examine the correlation between CGPA, academic pressure, and financial concerns. A heatmap of the correlation matrix for these three variables will help quantify the strength and direction of relationships

```python
# Correlation heatmap for CGPA, academic_pressure, and financial_concerns
corr_matrix = df[['cgpa', 'academic_pressure', 'financial_concerns']].corr()
plt.figure(figsize=(5,4))
sns.heatmap(corr_matrix, annot=True, cmap='YlGnBu', vmin=-1, vmax=1)
plt.title('Correlation: CGPA, Academic Pressure, Financial Concerns')
plt.show()
```

#### The heatmap above displays correlation coefficients (Pearson’s r) between academic performance (CGPA), academic pressure, and financial concerns. The values are all near 0, indicating very weak correlations. Specifically, CGPA has a slight negative correlation with academic pressure and financial concerns, but these relationships are not strong (the cells are close to 0). In practical terms, this suggests that higher academic pressure or financial worries do not strongly predict a student’s GPA in this dataset. Many students with high stress still achieve high grades, and some students with low stress struggle academically, implying other factors (such as individual ability, study habits, or support systems) play a significant role in academic success. 

#### For university administrators, this insight highlights that tackling academic pressure or financial stress in isolation may not directly boost GPA. However, reducing excessive stress and financial anxiety is still crucial for student well-being, even if the immediate impact on grades is limited. The lack of strong correlation also sets the stage for our modeling attempts: it foreshadows that predicting GPA from just these stress factors might be challenging.

### Modeling
To further explore the relationship between mental health factors and academic performance, we attempt two types of predictive modeling:
#### Regression: Can we predict a student's numerical CGPA based on their mental health and stress indicators?
#### Classification: Can we classify students into performance categories (Low/Medium/High CGPA) using those indicators?
These modeling exercises help quantify how well (or poorly) these factors explain academic outcomes. All models are kept simple and interpretable, as the goal is insight rather than deployment.

### Regression: Predicting CGPA from Stress Factors
#### Regression Setup: We use a linear regression model to predict students’ CGPA (as a numeric value) from a subset of stress-related features: academic pressure, financial concerns, and social relationships. We focus on these three factors because they are major categories of stress that the university might influence via policy (academic workload, financial aid, and campus social support). We split the data into 70% training and 30% testing to evaluate the model’s performance on unseen data. We also experiment with a polynomial regression (degree 2) to see if allowing a non-linear combination of these factors improves prediction.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Features for regression and target variable
features = ['academic_pressure', 'financial_concerns', 'social_relationships']
X = df[features]
y = df['cgpa']

# Train-test split (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Linear regression model
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
y_pred = linear_model.predict(X_test)

# Evaluate linear regression
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Linear Regression - MSE: {mse:.3f}, MAE: {mae:.3f}, R^2: {r2:.3f}")

# Polynomial regression model (degree 2)
poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train)
y_pred_poly = poly_model.predict(X_test_poly)

# Evaluate polynomial regression
poly_mse = mean_squared_error(y_test, y_pred_poly)
poly_mae = mean_absolute_error(y_test, y_pred_poly)
poly_r2 = r2_score(y_test, y_pred_poly)
print(f"Polynomial Regression (deg 2) - MSE: {poly_mse:.3f}, MAE: {poly_mae:.3f}, R^2: {poly_r2:.3f}")
```

#### Interpretation:
#### The linear model produced a Mean Absolute Error (MAE) of 0.88, meaning it was off by nearly one CGPA point on average — a large error on a 4-point scale. The R² score of -0.27 indicates the model performs worse than simply predicting the average CGPA for all students. The polynomial model (degree 2) showed slightly better metrics (MAE: 0.84, R²: -0.21) but still failed to capture any meaningful pattern.

#### Implication:
#### These results suggest that academic pressure, financial concerns, and social relationships alone are not strong predictors of CGPA. Academic performance likely depends on additional factors like prior achievement, motivation, and learning support. For stakeholders, this highlights the need for richer data and more advanced models if GPA prediction is a goal.

### Clustering
#### Besides prediction, we can explore unsupervised clustering to see if students naturally group into distinct profiles based on their mental health and stress factors. We apply K-Means clustering on three key features: academic pressure, financial concerns, and social relationships. The idea is to identify clusters of students who have similar stress and support profiles, which could help universities tailor interventions to different student segments. We standardized the features (so that one factor doesn’t dominate due to scale differences) and used an elbow method to choose an appropriate number of clusters. The elbow analysis suggested about 3 clusters as a reasonable choice. We then fit a K-Means model with 3 clusters, and each student is assigned a cluster label (0, 1, or 2).

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Select and scale features for clustering
X_clustering = df[['academic_pressure', 'financial_concerns', 'social_relationships']]
scaler = StandardScaler()
X_clustering_scaled = scaler.fit_transform(X_clustering)

# K-Means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(X_clustering_scaled)

# Display cluster centers (in scaled units for reference)
print("Cluster Centers (scaled features):\n", kmeans.cluster_centers_)
# and a quick look at how many students fall into each cluster
print("\nCluster membership counts:\n", df['cluster'].value_counts())
```

#### Now that we have cluster assignments, we can analyze the characteristics of each cluster. A useful way to interpret the clusters in context is to see how they correlate with academic performance. We create a bar chart to show the distribution of CGPA performance bands within each cluster. This will tell us, for example, if one cluster contains a higher proportion of low-performing students, or if another cluster is mostly high achievers.

```python
# Bar chart: CGPA band counts by cluster
plt.figure(figsize=(8,6))
sns.countplot(x='cluster', hue='cgpa_band', data=df, palette='Set2')
plt.title('Academic Performance by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Number of Students')
plt.legend(title='CGPA Band')
plt.show()
```

### Interpretation:

#### Cluster 2:  Dominated by Medium and High CGPA students. This group likely includes well-supported students with manageable stress levels.

#### Cluster 1:  Contains a higher ratio of high CGPA students. This cluster may represent the most academically at-risk group, potentially under high stress or lacking support.

#### cluster 0: Mostly Medium performers with few extremes. This group shows a mixed composition of performance, perhaps with a moderate number of high and low CGPA students. This cluster might represent students with moderate stress and support levels.

#### These clusters help identify student profiles. For example, Cluster 1 students may benefit from targeted support like mentoring or counseling, while Cluster 2 highlights what conditions support success. While clustering doesn’t imply causation, it helps universities prioritize support for students most at risk.

### Classification: Predicting CGPA Band from Stress Factors

We now attempt a classification model to predict a student's **CGPA performance band** (Low, Medium, High) using the following features:

- Academic Pressure
- Financial Concerns
- Social Relationships

This model helps us explore whether stress indicators can reliably classify students into academic performance categories. We use a **Decision Tree** for its simplicity and explainability.


```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Features and target
X_class = df[['academic_pressure', 'financial_concerns', 'social_relationships']]
y_class = df['cgpa_band']

# Split into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_class, y_class, test_size=0.3, random_state=42)

# Train a decision tree classifier
clf = DecisionTreeClassifier(random_state=42, max_depth=3)
clf.fit(X_train, y_train)

# Predict on test set
y_pred = clf.predict(X_test)

# Evaluation
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
print("\nClassification Report:\n")
print(classification_report(y_test, y_pred, zero_division=0))

# Confusion matrix
plt.figure(figsize=(6,4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt="d", cmap="YlOrRd",
            xticklabels=clf.classes_, yticklabels=clf.classes_)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

### Interpretation

The model achieved an accuracy of approximately **XX%** (replace with actual value). The classification report shows how well each CGPA band was predicted.

- **High performers** were more easily identified due to being the majority class.
- **Low and Medium bands** were harder to classify, likely due to overlapping stress profiles and fewer training examples.

The confusion matrix shows prediction counts for each class. This basic model shows that **stress features alone are not strong predictors** of academic band, but the results offer a starting point for early risk detection when combined with other data.
